# Ploting dengan Pandas

## 0. Import Library & Load data

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "tanggal": pd.date_range("2024-01-01", periods=10),
    "produk": ["A", "B", "A", "C", "B", "A", "C", "A", "B", "C"],
    "jumlah": [10, 5, 8, 6, 7, 9, 4, 11, 6, 5],
    "harga": [10000, 15000, 10000, 20000, 15000, 10000, 20000, 10000, 15000, 20000]
}

df = pd.DataFrame(data)
df["total"] = df["jumlah"] * df["harga"]
df
```

## 1. Line Plot

Menampilkan perubahan total penjualan dari hari ke hari.

```python
df.plot(
    x="tanggal",
    y="total",
    kind="line",
    marker="o",
    title="Tren Total Penjualan Harian"
)

plt.xlabel("Tanggal")
plt.ylabel("Total Penjualan")
plt.show()
```

## 2. Bar Plot

Membandingkan total pendapatan setiap produk.

```python
df.groupby("produk")["total"].sum().plot(
    kind="bar",
    title="Total Penjualan per Produk"
)

plt.xlabel("Produk")
plt.ylabel("Total Penjualan")
plt.show()
```

## 3. Horizontal Bar Plot

Menampilkan jumlah unit terjual tiap produk secara horizontal.

```python
df.groupby("produk")["jumlah"].sum().plot(
    kind="barh",
    title="Jumlah Produk Terjual"
)

plt.xlabel("Jumlah Terjual")
plt.ylabel("Produk")
plt.show()
```

## 4. Histogram

Menampilkan sebaran nilai total penjualan.

```python
df["total"].plot(
    kind="hist",
    bins=5,
    title="Distribusi Total Penjualan"
)

plt.xlabel("Total Penjualan")
plt.show()
```

## 5. Boxplot

Menampilkan distribusi dan outlier penjualan per produk.

```python
df.boxplot(
    column="total",
    by="produk"
)

plt.title("Variasi Total Penjualan per Produk")
plt.suptitle("")
plt.xlabel("Produk")
plt.ylabel("Total Penjualan")
plt.show()
```


## 6. Scatter Plot

Menampilkan hubungan antara jumlah produk terjual dan total pendapatan.

```python
df.plot(
    kind="scatter",
    x="jumlah",
    y="total",
    title="Hubungan Jumlah Terjual dan Total Penjualan"
)

plt.xlabel("Jumlah Terjual")
plt.ylabel("Total Penjualan")
plt.show()
```

## 7. Area Plot

Visualisasi kontribusi total penjualan tiap produk.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat contoh dataframe
data = {
    "produk": ["A", "B", "A", "C", "B", "C", "A"],
    "total": [100, 150, 200, 120, 180, 130, 170]
}

df = pd.DataFrame(data)

# Pastikan kolom total numerik
df["total"] = pd.to_numeric(df["total"], errors="coerce")

# Grouping data
df_grouped = df.groupby("produk")["total"].sum()

# Plot area
df_grouped.plot(
    kind="area",
    title="Akumulasi Total Penjualan per Produk"
)

plt.xlabel("Produk")
plt.ylabel("Total Penjualan")
plt.show()
```


## 8. Pie Chart

Menampilkan persentase kontribusi produk terhadap total penjualan.




```python
df.groupby("produk")["total"].sum().plot(
    kind="pie",
    autopct="%1.1f%%",
    title="Proporsi Penjualan Produk"
)

plt.ylabel("")
plt.show()
```
