# Ploting dengan seaborn


## 0. Import & Data Load

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Pengaturan Tema
sns.set_theme(
    style="whitegrid",
    context="notebook",
    palette="pastel"
)

# Load Dataset (Dataset Bawaan Seaborn)
df = sns.load_dataset("tips")
print(df.head())
```

## 1. Line Plot

Kegunaan:
- Menunjukkan tren perubahan nilai terhadap variabel berurutan
- Cocok untuk time series atau data numerik bertingkat

```python
sns.lineplot(data=df, x="size", y="total_bill")
plt.title("Rata-rata Total Bill berdasarkan Ukuran Meja")
plt.show()
```

## 2. Scatter Plot

Kegunaan:
- Melihat hubungan dua variabel numerik
- Mengamati pola, klaster, dan outlier

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", style="smoker")
plt.show()
```

## 3. Histogram

Kegunaan:
- Mengetahui bentuk distribusi data
- Mendeteksi skewness dan sebaran nilai

```python
sns.histplot(df["total_bill"], kde=True)
plt.show()
```


## 4. Boxplot

Kegunaan:
- Membandingkan distribusi antar kategori
- Mendeteksi outlier

```python
sns.boxplot(data=df, x="day", y="total_bill")
plt.show()
```


## 5. Violin Plot

Kegunaan:
- Kombinasi boxplot dan distribusi density
- Menunjukkan bentuk distribusi lebih detail

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

sns.violinplot(data=df, x="day", y="total_bill", hue="sex", split=True)

plt.show()
```


## 6. Bar Plot

Kegunaan:
- Menampilkan nilai agregat (rata-rata, median, dll)
- Membandingkan performa antar kategori

```python
sns.barplot(data=df, x="day", y="tip", estimator="mean", errorbar=None)
plt.show()
```

## 7. Heatmap

Kegunaan:
- Melihat hubungan linear antar variabel numerik
- Identifikasi variabel yang saling berkorelasi

```python
corr = df.select_dtypes("number").corr()
sns.heatmap(corr, annot=True)
plt.show()
```

## 8. Pairplot

Kegunaan:
- Exploratory Data Analysis (EDA) cepat
- Melihat hubungan banyak variabel sekaligus

```python
sns.pairplot(df, hue="sex")
plt.show()
```

## 9. Countplot

Kegunaan:
- Menghitung frekuensi data kategori
- Membandingkan jumlah antar kelompok

```python
sns.countplot(data=df, x="day", hue="smoker")
plt.show()
```
